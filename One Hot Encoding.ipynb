# Dummie Variables

```python
import pandas as pd 
df = pd.read_csv("../CSVs/one_hot_encoding.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged_df = pd.concat([df,dummies],axis='columns')
merged_df
```

```python
merged_df = merged_df.drop(['town','west windsor'],axis='columns')
merged_df
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = merged_df.drop(['price'],axis='columns')
```

```python
Y = df['price']
```

```python
model.fit(X,Y)
```

```python
model.predict([[2000,1,0]])
```

```python
model.score(X,Y)
```

# One Hot Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() 
```

```python
df2 = df
df2.town = le.fit_transform(df2.town)
df2
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(drop='first', sparse=False)
encoded_data = encoder.fit_transform(df2[['town']])
feature_names = encoder.get_feature_names_out(input_features=['town'])
encoded_df = pd.DataFrame(encoded_data, columns=feature_names)
final_data = pd.concat([df2, encoded_df], axis=1).drop('town', axis=1)
final_data
```

```python
X2 = final_data.drop(['price'],axis='columns')
Y2 = final_data['price'] 
```

```python
model.fit(X2,Y2)
```

```python
model.predict([[1,0,2800]])
```

```python
model.score(X2,Y2)
```
